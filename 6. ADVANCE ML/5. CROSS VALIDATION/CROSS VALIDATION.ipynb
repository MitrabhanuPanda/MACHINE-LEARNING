<h1 style="text-align : center"> <font color="red" size=8>CROSS VALIDATION </h1>

## <font color="dark blue">WHAT IS CROSS VALIDATION?
- It is a technique used in Machine Learning to evaluate the performance of our model on unseen data.
- eg:
    
> - Imagine you're baking a cake. You have a recipe, but you're not sure if it's the best one. Instead of just baking one cake and tasting it, you divide the ingredients into 5 parts. You use 4 parts to bake 4 different cakes, and you taste each one. Then, you use the 5th part to bake a final cake, using the best recipe from the first 4

## <font color="dark blue">WHY WE USE CROSS VALIDATION?
- To prevent overfitting issue.
- To provide more realistic estimate of model's generalization performance.

## <font color="dark blue">TYPES OF CROSS VALIDATION?

## <font color="blue">a. HOLD OUT STRATEGY
- In this technique, we will pass our hyperparameters in a loop which gives both train & test score & from this score which give the best score we choose that hyperparameter for our final model.

### <font color="orange">LIMITATION
- Data leakage is a big issue.

- Here k=10 is the best hyperparameter value which give the best model.

## <font color="blue">b. CROSS VALIDATION TECHNIQUE
- In this technique, we will split the data into Training & Testing. We again Split the Training set into Training & Validation.
- We train the algorithm using Training set & generate validation score then test the model using Testing set.

### <font color="orange">LIMITATION
- This technique can lead to underfitting since the training data is reducing.

## <font color="blue">c. K-FOLD CROSS VALIDATION TECHNIQUE
- In this technique, We take whole dataset split into __k__ fold(k subset), then we train the algorithm in all the subset but leave one subset for evaluation of trained model.
- We train the model __k__ times for different differnet subsets.


### <font color="orange">LIMITATION
- Training time is increses.

## <font color="green">HOW K-FOLD CROSS VALIDATION TECHNIQUE WORKS?

### <font color="green">STEPS
1. Choose a number of folds, usually `5` or `10`.
2. Split the dataset into equal parts called `folds`.
3. Choose `k-1` folds as the training set, and the remaining fold as the test set.
4. Train the model on the training set.
5. Validate on the test set.
6. Repeat steps 3-6,  `k` times, each time using a different fold as the test set.
7. Average the results from `STEP-6` to get the final score. 

# <font color="brown">CODING FOR K-FOLD CROSS VALIDATION TECHNIQUE

### <font color="GREEN">IMPORT LIBARRIES

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
```

### <font color="GREEN">IMPORT DATASET & SPLIT INTO X & y

```python
df=load_iris()
X,y = df.data, df.target
```

### <font color="GREEN">CREATE VARIABLE OF ALGORITHM CLASS

```python
dt=DecisionTreeClassifier()
```

### <font color="GREEN">DEFINE K-FOLD FOR CROSS VALIDATION

```python
num_folds=5
kf=KFold(n_splits=num_folds, shuffle=True, random_state=42)
```

### <font color="GREEN">PERFORMING K-FOLD CROSS VALIDATION

```python
cros_val_res=cross_val_score(dt,X,y,cv=kf)
```

### <font color="GREEN">EVALUATE THE MODEL PERFORMANCE

```python
print(f'Cross-Validation Results (Accuracy): {cros_val_res}')
print(f'Mean Accuracy: {cros_val_res.mean()}')
```

## <font color="purple">PARAMETERS OF CROSS_VAL_SCORE FUNCTION
__1. estimator__:
- The machine learning model to be evaluated. This can be any estimator object from scikit-learn.

__2. X__:
- The input features or independent variables.

__3. y (optional)__:
- The target variable or dependent variable.

__4. groups (optional)__:
- Optional group labels for the data. 
- Used for group-based cross-validation.

__5. scoring (optional)__:
- The metric to evaluate the model's performance. 
- Common options include accuracy, precision, recall, f1, roc_auc'\, etc.

__6. cv (optional)__:
- The cross-validation strategy. 
- It can be an integer (for k-fold cross-validation), a cross-validation object, or a generator.

__7. n_jobs (optional)__:
- The number of CPU cores to use for parallel computation.

__8. verbose (optional)__:
- Controls the verbosity of the output.

__9. params (optional)__:
- A dictionary of parameters to pass to the estimator.

__10. pre_dispatch (optional)__:
- Controls the number of jobs dispatched to the executor.

__11. error_score (optional)__:
- Value to return when a fit fails.
