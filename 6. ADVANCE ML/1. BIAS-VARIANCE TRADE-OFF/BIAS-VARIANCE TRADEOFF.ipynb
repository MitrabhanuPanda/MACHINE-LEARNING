## <font color="green">NOTE

## <font color="dark blue">WHAT IS OVERFITTING?
- Overfitting is a common problem in machine learning where a model becomes too complex and fits the training data too closely. This results in the model performing poorly on new, unseen data

## <font color="dark blue">WHY DOES OVERFITTING HAPPEN?
- Complex Models
- Noise in Data
- Insufficient Training Data

## <font color="dark blue">WHAT IS UNDERFITTING?
- Underfitting occurs when a machine learning model is too simple to capture the underlying patterns in the training data. As a result, the model performs poorly on both the training and testing data.

## <font color="dark blue">WHY DOES UNDERFITTING HAPPEN?
- Simpler Models
- Insufficient Training Data

## <font color="dark blue">HOW TO OVERCOME UNDERFITTING ISSUE?
- Increase Model Complexity
- Reduce Regularization
- Increase in Training Data
- Feature Enginerring
- Algorithm Selection

### 

## <font color="dark blue">WHAT IS BIAS?
- Bias refers to a systematic error that occurs when a model makes incorrect assumptions about the data. It's like a model consistently missing the mark in a particular direction
 
- The bias is known as the difference between the prediction of the values by the Machine Learning model and the correct value. 


- Being high in biasing gives a large error in training as well as testing data


- The bias error is an error from erroneous assumptions in the learning algorithm.


## <font color="dark blue">HOW TO REDUCE BIAS?
- Collect More Data.
- Use more complex models.
- Feature Engineering

## <font color="dark blue">WHAT IS VARIANCE?
- The variability of model prediction for a given data point which tells us the spread of our data is called the variance of the model.


- The variance is an error from sensitivity to small fluctuations in the training set.


- In machine learning, variance refers to the sensitivity of a model's predictions to changes in the training data. A model with high variance is highly sensitive to small fluctuations in the training data


## <font color="dark blue">HOW TO REDUCE VARIANCE?
- Increase in Training Data
- Regularization
- Feature Selection
- Ensemble Methods

## <font color="dark blue">WHAT IS PARAMETER?
- Parameters are learned by the model itself during the training process.
- They are part of the model's internal state & are directly involved in making predictions.
- eg: $\rightarrow$ Weights & Biases in Neural Network, Coefficient in Linear Regression.
    
## <font color="dark blue">WHAT IS HYPERPARAMETER?  
- Hyperparameters are set by ML Engineer / Data Scientist (Users) before the training process begins.
- They influence how the model learns & its overall behavior.
- eg $\rightarrow$ Learning Rate, Batch Size, Number of Epochs, Max Depth

FEATURES | PARAMETER | HYPERPARAMETER
--------- | ---------- | -------------
__SOURCE__ | Learned from Data | Set by ML Engineer / Data Scientist (Users)
__ROLE__ | Directly involved in Predictions | Control the learning Process
__OPTIMIZATION__ | Optimized during training | Tuned manually or using techniques like Grid Search or Random Search

## <font color="dark blue">WHAT IS PARAMETRIC ALGORITHMS?
- Algorithms which generates models with countable finite parameters are called __PARAMETRIC ALGORITHMS__.

## <font color="dark blue">WHAT IS NON-PARAMETRIC ALGORITHMS?
- Algorithms which generates models with no parameters are called __NON-PARAMETRIC ALGORITHMS__.
