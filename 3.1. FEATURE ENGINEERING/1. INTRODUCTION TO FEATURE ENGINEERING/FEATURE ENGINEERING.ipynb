<h1 style="text-align : center"> <font color="red" size=7>FEATURE ENGINEERING </h1>

## <font color="dark blue">WHAT IS FEATURE ENGINEERING?<font color/>
- Feature Engineering is the process of using domain knowldge to extract features from raw data.
- These features can be used to improve the performance of Machine learning Algorithms.

## <font color="dark blue">WHY FEATURE ENGINEERING IS IMPORTANT?<font color/>
- Feature engineering is important because it helps machine learning models learn better by creating meaningful and relevant features from raw data. 
    
    
- Well-engineered features improve model accuracy, make training more efficient, and help models generalize well to new data. 
    
    
- It involves techniques like handling missing values, encoding categorical data, scaling numerical features, and creating new features from existing ones. 
    
    
- Good feature engineering can sometimes be more impactful than choosing a complex model, leading to better predictions and performance.
