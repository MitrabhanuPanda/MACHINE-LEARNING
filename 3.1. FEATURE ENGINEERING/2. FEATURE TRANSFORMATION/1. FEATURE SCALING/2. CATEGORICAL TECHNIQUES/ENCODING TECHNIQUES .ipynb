<h1 style="text-align : center"> <font color="red" size=7>ENCODING TECHNIQUES </h1>

## <font color="dark blue">WHAT IS ENCODING TECHNIQUES?<font color/>
- It is machine learning techniques where we convert the categorical data to numerical data.
- Because Machine learning Algorithms only know/understand the Numeric data.

## <font color="dark blue">TYPES OF ENCODING TECHNIQUES?<font color/>

<h1 style="text-align : center"> <font color="purple" size=7>ORDINAL ENCODING </h1>

## <font color="dark blue">WHAT IS ORDINAL ENCODING?<font color/>
- It has order of the data in categorical features.
- eg: Grade, Devision

<h1 style="text-align : center"> <font color="purple" size=7>ONE-HOT ENCODING </h1>

## <font color="dark blue">WHAT IS ONE-HOT ENCODING?<font color/>
- It is a machine learning technique where it convert categorical values to numerical values.
- It know about the unique values in a categorical feature, then create features, then fill the value with __0__ & __1__.
- __1__ means The specific feature is in this record, __0__ means The specific feature is not in this record.
