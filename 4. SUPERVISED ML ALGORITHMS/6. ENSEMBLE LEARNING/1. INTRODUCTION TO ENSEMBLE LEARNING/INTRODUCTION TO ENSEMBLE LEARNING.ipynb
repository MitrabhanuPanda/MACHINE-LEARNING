<h1 style="text-align : center"> <font color="red" size=8>INTRODUCTION TO ENSEMBLE LEARNING</h1>

## <font color="dark blue">WHAT IS ENSEMBLE LEARNING?
- Ensemble means __Grouping / Collection of things__.
- Ensemble Learning is a machine learning paradigm where multiple models are trained to solve the same problem & their predictions are combined to produce a more accurate & robust result.
- It's like seeking advice from multiple experts to make a more informed decision.

## <font color="dark blue">TYPES OF ENSEMBLE LEARNING?
__1. VOTING ENSMEBLE__:
- Multiple models are trained on the same data & then majority voting is used to predict the result.
    
   
__2. STACKING ENSEMBLE__:
- Multiple models are trained on the same data & their predictions are used as input to a Meta-model, which makes the final prediction.
    
__3. BAAGGING ENSEMBLE__:
- BAGGING Stands for __`Boostrap Aggregation`__.
- Multiple models are trained on different subsets of the training data, created by random sampling with replacement.
- The final prediction is the average or majority vote of all models.
    
__4. BOOSTING ENSEMBLE__:
- Models are trained sequentially with each model focusing on correcting the errors of the previous ones.

## <font color="dark blue">BENEFITS OF ENSEMBLE LEARNING?
- Improve predictive performance
- Reduce Overfitting
- Increase Robustness.

## <font color="dark blue">WHY ENSEMBLE APPROACH ARE SO POPULAR?
- Because we train multiple model using multiple dataset, so that the model are learn different patterns from data.
- For that reason it give better prediction than a single model.

## <font color="dark blue">USES OF ENSEMBLE APPROACH
- Anomaly Detection
- Customer Churn Prediction
- Credit Scoring
- Climate Prediction
- Time Series Forecasting
