###### <h1 style="text-align : center"> <font color="red" size=8>LINEAR REGRESSION </h1>

## <font color="dark blue">WHAT IS LINEAR REGRESSION?
- It comes under  __Boundary Base Approach__.
- It is a Supervised ML Algorithm which works in Regression task when the target variable is Continuous.
- __`Find a line/plane/hyperplane that best fit on the given data.`__
    - __`BEST FIT means LESS ERROR`__
- Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables. 
    - It assumes a linear relationship between the variables, meaning the change in the dependent variable is proportional to the change in the independent variable(s).

## <font color="dark blue">HOW TO REPRESENT A BOUNDARY?
- If the Boundary is a Straight Line, we can use Equation of Line to represent that boundary.
$$\large 2D \ Equation \ of \ Boundary \ \Rightarrow y=mx+c$$
$$where \ , $$
$$  c \ = \ Intercept \ w.r.t \ \ y-axis$$
$$ m \ = \ Slope  \ w.r.t \ \ x-axis$$
$$ m \ =  \tan\theta $$
$$ m \ and \ c \ are \ Parameters \ of \ Line. $$

## <font color="dark blue">GENERALIZATION OF 2D EQUATION OF LINE
$$ ax+by=c$$
$$ where \ ,$$
$$ a \ , \ c \ are \ coefficient \ of \ line$$
- If you want to know about the Slope & Intercept then you have to rewrite the equation in $y=mx+c$ form.
$$ y=\frac{-a}{b}+\frac{-c}{b}$$
$$ where \ ,$$
$$ Slope \ w.r.t \ x-axis \rightarrow \frac{-a}{b}$$
$$ Intercept \ w.r.t \ y-axis \rightarrow \frac{-c}{b}$$

## <font color="dark blue">GENERALIZATION EQUATION OF LINE in D-DIMENSION
$$d \ Dim \rightarrow \sum_{i=1}^{n}w_ix_i+w_o$$
- Slopes $ \rightarrow d-1 $
- Intercept $ \rightarrow 1 $
    
$$ \prod_{d} : W^Tx+W_o = 0 \ (D-Dimension \ Hyperplane \ Equation)  $$

## <font color="blue">SLOPE
- It is the trigonometric `tan` of angle created between the line & the x-axis.
- Rate of Change $\rightarrow$ How much __y__ will change with 1 unit change in __x__.

## <font color="blue">INTERPRETATION OF EQUATION OF A LINE
$$ y=mx+c $$
- m $\rightarrow$ with 1 unit change in __x__, How much __y__ will change.
- c $\rightarrow$ Even if all the input are __0__, what will be the value of __y__.

## <font color="dark blue"> OPTIMIZATION EQUATION OF LINEAR REGRESSION
$$ \large \overrightarrow{W}^* \ , \ \overrightarrow{W}_{o}^* = arg_{\overrightarrow{W},\overrightarrow{W}_{o}} min \{ \sum_{i=1}^{n} (y_i - \widehat{y_i})^2 \}$$
    
    
 $$ \large \overrightarrow{W}^* \ , \ \overrightarrow{W}_{o}^* = arg_{\overrightarrow{W},\overrightarrow{W}_{o}} min \{ \sum_{i=1}^{n} (y_i-(\frac{-w_i}{w2}x_i - \frac{w_0}{w_2}))^2 \}$$  

## <font color="dark blue"> TYPES OF LINEAR REGRESSION

### <font color="blue"> 1. SIMPLE LINEAR REGRESSION
- If the dataset having only one independent feature then it comes under __SIMPLE LINEAR REGRESSION__.

### <font color="blue"> 2. MULTIPLE LINEAR REGRESSION
- If the dataset having more than one independent feature then it comes under __MULTIPLE LINEAR REGRESSION__.

## <font color="dark blue"> DIFFERENCE BETWEEN COST FUNCTION & LOSS FUNCTION

## <font color="blue">Loss Function:
- Measures the error for a single data point.   
- Calculates the difference between the predicted value and the actual value for a specific data instance.   
- Used to evaluate the model's performance on a single data point.   


## <font color="blue">Cost Function:
- Measures the average error across the entire training dataset.   
- Calculates the average of the loss functions for all data points in the training set.   
- Used to evaluate the overall performance of the model on the entire training set.



## <font color="dark blue"> ADVANTAGES
- Simplicity
- Interpretability
- Efficiency

## <font color="dark blue"> DISADVANTAGES
- Linearity Assumption
- Sensitive to Outliers
- Overfitting

### <font color="blue">1. USING SKLEARN

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\DUDUL DS\CAMPUSX\ML\LINEAR REGRESSION\DATASET\student_scores.csv")
```

```python
X = df[['Hours']]
y = df['Scores']
```

```python
# Train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print('y intercept: ',regressor.intercept_)
```

```python
print('slope: ',regressor.coef_)
```

```python
y_test_pred = regressor.predict(X_test)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})
temp_df
```

```python
from sklearn import metrics

print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_test_pred))

print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_test_pred))

print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```

```python
metrics.median_absolute_error(y_test, y_test_pred)
```

```python
print(metrics.r2_score(y_test, y_test_pred))
```

```python
r2 = metrics.r2_score(y_test, y_test_pred)
n = len(y_test)
k = X_test.shape[1]

r2_adj = 1 - (1-r2)*(n-1)/(n-k-1)

print(r2_adj)
```

### <font color="blue">2. USING STATSMODELS

```python
import statsmodels.api as sm
```

```python
# Add a constant to get an intercept
X_train_sm = sm.add_constant(X_train)

# Fit the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()
```

```python
lr.params
```

```python
print(lr.summary())
```

## <font color="dark blue"> ASSUMPTION OF LINEAR REGRESSION
- Linearity Assumption (between input & output)
- Residuals should follow __Normal Distribution__ with __Zero Mean__.
- Residuals should be independent of each other & follow __identical distribution (IID)__.
- Residuals should follow __constant variance (HOMOSCEDASTICITY)__.

    
- Multicollinearity Check
- Independence of Observations/Rows.

## <font color="blue">WHAT IS RESIDUAL?
- In machine learning, a residual is the difference between the actual value of a dependent variable and the predicted value from a model. 
- In simpler terms, it's the error made by the model for a particular data point

### RESIDUAL ANALYSIS ON TRAINING DATA
- Distribution of the Residuals - Distribution should be Normal (Gaussian) with Zero Mean.
- IID -  Check for patterns in Residuals - There should exist no pattern.
- Homoscedasticity - The variance of residaul is the same for any value of X.

## <font color="purple"> PARAMETERS OF LINEAR REGRESSION CLASS

__1. fit_intercept (default=`True`)__:
- Determines whether to calculate the intercept (y-intercept) of the linear model.
- If `True`, the model will have an intercept term.
- If `False`, the model will __pass__ through the origin.


__2. copy_X (default=`True`)__:
- Determines whether to copy the input data (X) before fitting the model.
- If `True`, a copy is made, which can be useful for certain operations.
- If `False`, the original data is used, which can be more memory-efficient.


__3. n_jobs (default=`None`)__:
- Specifies the number of CPU cores to use for parallel computation.
- If `None`, all available cores are used.
- This parameter is useful for speeding up computations on large datasets.


__4. positive (default=`False`)__:
- Restricts the coefficients to be positive.
- This can be useful in certain applications where the coefficients are expected to be non-negative.
