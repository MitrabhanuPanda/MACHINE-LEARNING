## <font color="dark blue">WHAT IS LOGISTIC REGRESSION?
- It comes under __Boundary Base Approach__.
- It is a __Supervise ML__.
- It is used to do Classification Task.
- It is used to do only __Binary Classification__.
- __`FIND A MODEL THAT BEST SEPARATE +ve's from -ve's.`__
- Logistic Regression is a statistical method used in machine learning for classification tasks. Unlike linear regression, which predicts a continuous numerical value, logistic regression predicts the probability of a binary outcome (e.g., 0 or 1, yes or no).

## <font color="dark blue">HOW TO REPRESENT IT MATHEMATICALLY?
$$ \large \ m^* \ , \ \ c^* = arg_{m,c} max \ \{ Correct \ Classification \}$$
    
$$ \Rightarrow \large \ m^* \ , \ \ c^* = arg_{m,c} max \ \{ \sum_{i=1}^{n} (actual_i \ * \ predicted_i) \}$$
    
$$ \Rightarrow \large \ m^* \ , \ \ c^* = arg_{m,c} max \ \{ \sum_{i=1}^{n} (y_i \ * \ \widehat{y_i}) \}$$

$$ \Rightarrow \large \ m^* \ , \ \ c^* = arg_{m,c} max \ \{ \sum_{i=1}^{n} (y_i \ * \ (mx_i+c)) \}$$
    
 $$ \large \overrightarrow{W}^* \ , \ \overrightarrow{W}_{o}^* = arg_{\overrightarrow{W},\overrightarrow{W}_{o}} max \{ \sum_{i=1}^{n} (y_i(w^Tx_i+w_0)) \}$$  
    
- It is the Optimization Equation of Logistic Regression.
    
- It is also known as __SIGNED DISTANCE__.
    
$$ - \infty \ \leq \  SIGNED \ DISTANCE \ \leq \ +\infty $$
    
- It is affected by Outliers. So to overcome this issue, we use Sigmoid Function.

## <font color="dark blue">WHAT IS LOGISTIC FUNCTION / SIGMOID FUNCTION?
- The sigmoid function is a mathematical function used to map the predicted values to probabilities.
    
- It maps any real value into another value within a range of 0 and 1. The value of the logistic regression must be between 0 and 1, which cannot go beyond this limit, so it forms a curve like the `S` form.
    
- The S-form curve is called the Sigmoid function or the logistic function.
    
- In logistic regression, we use the concept of the threshold value, which defines the probability of either 0 or 1. Such as values above the threshold value tends to 1, and a value below the threshold values tends to 0.
    
$$ \large sigmoid \ function \  \sigma (x) \ = \frac{1}{1+e^{-x}} $$
    
$$ \Rightarrow \large sigmoid \ function \  \sigma (x) \ = \frac{e^x}{e^x+1} $$
    
- It reduces the impact of Outlier by rescaling the Signed Distance in the range of 0 to 1.
    
    
$$ 0 \ \leq \ \sigma  \ \leq \ 1 $$
    
- $ \sigma $ is in the range of 0 to 1, for that we can be interpreted as `PROBABILITY`.

# <font color="dark blue">SOME IMPORTANT NOTE ON LOGISTIC REGRESSION

## <font color="blue">LOGIT ODDS
- In statistics and machine learning, logit odds is a mathematical transformation of probability. It's the natural logarithm of the odds ratio
- Regession models the log of odds of an event.
$$\large odds \ = \ \frac{p}{1-p}$$
$$ p \rightarrow Probability \ of \ an \ event \ happening $$
$$\large p \ = \ \frac{odds}{1+odds}$$
    
$$ p \ ranges \ from \ 0 \ to \ 1.$$

## <font color="blue">DIFFERENCE BETWEEN PROBABILITY & LIKELIHOOD
FEATURE | PROBABILITY | LIKELIHOOD
----------- | ---------- | ---------
__FOCUS__ | Future Events | Past Parameters
__PERSPECTIVE__ | Fixed parameters, variable outcomes | Fixed outcomes, variable parameters
__CALCULATION__ | Based on known probabilities | Based on observed data.

```python
y_pred_prob = classifier.predict_proba(X_test)
metrics.log_loss(y_test, y_pred_prob)
```

## <font color="purple">PARAMETERS OF LOGISTIC REGRESSION CLASS

__1. penalty (default=`l2`)__:
- Specifies the regularization technique.
    - __l2__: L2 regularization (Ridge regression), adds a penalty term to the loss function to prevent overfitting.
    - __l1__: L1 regularization (Lasso regression), adds a penalty term to the loss function to encourage sparsity in the model.
    - __elasticnet__: Elastic Net regularization, combines L1 and L2 regularization.

__2. dual (default=`False`)__:
- Determines whether to use the dual formulation of the optimization problem.
- Generally, the primal formulation is more efficient, so this parameter is usually set to False.

__3. tol (default=`0.0001`)__:
- Tolerance for stopping the optimization algorithm.
- A smaller tolerance leads to more precise solutions but may require more iterations.

__4. C (default=`1.0`)__:
- Inverse of the regularization strength.
- A smaller C value indicates stronger regularization.

__5. fit_intercept (default=`True`)__:
- Determines whether to calculate the intercept (bias) term.
    - If `True`, the model will have an intercept term.
    - If `False`, the model will pass through the origin.

__6. intercept_scaling (default=`1`)__:
- Scales the intercept term.
- This can be useful for numerical stability.

__7. class_weight (default=`None`)__:
- Weights associated with classes in the form {class_label: weight}.
- This can be used to address imbalanced datasets.

__8. random_state (default=`None`)__:
- Seed for the random number generator.
- Used for reproducibility.

__9. solver (default=`lbfgs`)__:
- Algorithm to use for optimization.
- Other options include `newton-cg`, `liblinear`, and `sag`.

__10. max_iter (default=`100`)__:
- Maximum number of iterations for the solver.


__11. multi_class (default=`deprecated`)__:
- Strategy to use for multi-class classification.
- This parameter is deprecated and will be removed in future versions.

__12. verbose (default=`0`)__:
- Controls the verbosity of the output.

__13. warm_start (default=`False`)__:
- Reuse the solution of the previous call to fit as initialization, potentially saving computation time.

__14. n_jobs (default=`None`)__:
- Number of CPU cores to use for parallel computation.

__15. l1_ratio (default=`None`)__:
- The mixing parameter for Elastic-Net regularization.

## <font color="dark blue">LOGISTIC REGRESSION FOR MULTICLASS CLASSIFICATION SETUP
- Logistic regression also handle Multiclass classification problem, it is known as __Multinomial Logistic Regression__.

## <font color="dark blue">HOW IT WORKS?

### <font color="blue">1. One vs One (OvO):
- Suppose we have classes like A,B,C. 
- Each class have 100 Datapoint.
- Total Datapoint is 300.
- It builds models like:
    
    $MODEL1 : A \ vs \ B \rightarrow 200 \ Datapoints$
    
    $MODEL2 : A \ vs \ C \rightarrow 200 \ Datapoints$
    
    $MODEL2 : B \ vs \ C \rightarrow 200 \ Datapoints$
    
    
- It creates $n*C_2$ models.
    
    - n $\rightarrow$ No. of Classes
    - eg: $\rightarrow$ $3*C_2 = \frac{3!2!}{(3-1)!} \Rightarrow \frac{3*2*1}{2!} \Rightarrow \frac{6}{2*1} \Rightarrow \frac{6}{2}=3 \ Models$ 
   
    
- __`For each model training no need to perform operations on whole data.`__
    
    
### <font color="blue">2. One vs Rest (OvR):
- Suppose we have classes like A,B,C. 
- Each class have 100 Datapoint.
- Total Datapoint is 300.
- It builds models like:
    
     $MODEL1 : A \ vs \ B \ vs \ C \rightarrow 300 \ Datapoints$
    
    $MODEL2 : B \ vs \ C \ vs \ A \rightarrow 300 \ Datapoints$
    
    $MODEL2 : C \ vs \ A \ vs \ B \rightarrow 300 \ Datapoints$
    
    
- __`For each model training need whole data.`__
   
- __`Model is less`__
