## <font color="dark blue">DISTANCE METRIX
### <font color="blue"> 1. EUCLIDEAN DISTANCE
- It's the straight-line distance between two points in Euclidean space.
$$ \large dist( \overrightarrow{p},\ \overrightarrow{q}) = [\sum_{i=1}^{d}(p_i - q_i)^2]^\frac{1}{2}$$
    
### <font color="blue"> 2. MANHATTAN DISTANCE
- It's the distance between two points measured along axes at right angles.
- It is generally used when we are interested in the total distance traveled by the object instead of the displacement
$$ \large dist( \overrightarrow{p},\ \overrightarrow{q}) = [\sum_{i=1}^{d} abs(p_i - q_i)^1]^1 $$
    
### <font color="blue"> 3. MINKOWSKI DISTANCE
- It's the generalizable equation for these above distances.
$$ \large dist( \overrightarrow{p},\ \overrightarrow{q}) = [\sum_{i=1}^{d} abs(p_i - q_i)^k]^\frac{1}{k}$$

## <font color="dark blue">WHAT IS KNN ALGORITHM?
- KNN stands for __K-NEAREST NEIGHBOUR__.
- It is a supervised ML works both for Classification & Regression.
- It is a __Distance Based Approach__.
- It also known as __Lazy Learner__.
    - Because It do not build a model in the training time. The algorithm memorizes all the training dataset.
- It is a __Non-Parametric__.
    - Because it do not make assumption about the data it is analyzing.

## <font color="dark blue">HOW KNN ALGORITHM WORKS?
- Given $X_q$ (i.e. an unseen query point).

- Find the distance between $X_q$ & each datapoints in the x_train (i.e. historical data).

- Figure out the datapoints which are closest (most similar) to $X_q$.

- Which class having more closet datapoints then we take this class as the Result (Majority Voting).

- In Regression problem, we take mean of corresponding y values of KNN.

## <font color="dark blue">HOW TO CHOOSE THE K VALUE?
- K $\rightarrow$ How many closest neighbour?
- By default, in sklearn k=5.
- If the target feature is Binary that means cardinality=2, then we take k= odd number.
- If the target feature is Multiclass that means cardinality>2, then we take k=  odd/even number.

## <font color="orange">CODE OF KNN ALGORITHM

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df=pd.read_csv(r"D:\DUDUL DS\CAMPUSX\ML\KNN ALGORITHM\Iris.csv")
df.head()
```

```python
df.drop(columns="Id",inplace=True)
```

```python
df.head()
```

```python
df["Species"].value_counts()
```

```python
df["Species"]=df["Species"].apply(lambda x: 0 if x=="Iris-setosa" else (1 if  x=="Iris-versicolor" else 2) )
df["Species"].value_counts()
```

```python
X=df.drop(columns="Species")
y=df["Species"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=40)
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(X_train,y_train)
```

```python
y_test_predict=knn.predict(X_test)
y_test_predict
```

```python
accuracy_score(y_test,y_test_predict)
```

```python
confusion_matrix(y_test,y_test_predict)
```

## <font color="purple">PARAMETERS OF KNN ALGORITHM CLASS

__1. n_neighbors__:
- The number of neighbors to consider when making a prediction. A higher value can lead to smoother decision boundaries but might be computationally more expensive.

__2. weights__:
- Determines how to weight the influence of neighboring points.
    - __uniform__: All neighbors have equal weight.
    - __distance__: Neighbors closer to the query point have more influence.

__3. algorithm__:
- The algorithm used to compute the nearest neighbors.
    - __auto__: Lets the algorithm choose the most appropriate algorithm based on the data.
    - __ball_tree__: Uses a Ball Tree algorithm.
    - __kd_tree__: Uses a KD Tree algorithm.
    - __brute__: Uses a brute-force search.

__4. leaf_size__:
- Leaf size passed to BallTree or KDTree. This can affect the speed of the construction and query, as well as the memory required to store the tree.  

__5. p__:
- Power parameter for the Minkowski metric.
    - When p=1, it's equivalent to Manhattan distance.
    - When p=2, it's equivalent to Euclidean distance.

__6. metric__:
- The distance metric used to compute distances between data points.
    - `minkowski` is the default metric, which includes Euclidean and Manhattan distances as special cases.

__7. metric_params__:
- Additional keyword arguments for the distance metric.

__8. n_jobs__:
- The number of parallel jobs to run for neighbors search.
