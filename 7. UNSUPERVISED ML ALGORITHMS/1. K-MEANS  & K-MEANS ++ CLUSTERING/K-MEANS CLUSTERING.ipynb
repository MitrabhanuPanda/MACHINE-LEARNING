## <font color="dark blue">WHAT IS CLUSTERING?
- Clustering is a technique to group similar points together.

## <font color="dark blue">APPLICATIONS OF CLUSTERING
- Customer Segmentation
- Image Segmentation
- Document Clustering
- Anomaly Detection

## <font color="dark blue">WHAT IS K-MEANS CLUSTERING?
- It is a Unsupervised ML Algorithm where we group unlabeled data into a specified number of clusters.

## <font color="dark blue">HOW K-MEANS CLUSTERING WORKS?
- Identify the __`k`__ Clusters.


- __STEP-1__ : $\rightarrow$ __INITIALIZATION STEP__
    - Randomly initialize __`k`__ centroid.
    

- __STEP-2__ : $\rightarrow$ __ASSIGNMENT STEP__
    - Assign each datapoints to its nearest Centroid.
    
    
- __STEP-3__ : $\rightarrow$ __CALCULATION STEP__
    - Recalculate the Centroids.
    $$ \large Centroid_k = \frac{\sum_{x_i  \epsilon \ set_k}  x_i}{|set_k|}$$
    
- __STEP-4__ : $\rightarrow$ __REPETATION STEP__
    - Repeat `STEP-2` & `STEP-3` until convergence.

## <font color="dark blue">HOW K-MEANS ++ CLUSTERING WORKS?
- Identify the __`k`__ Clusters.


- __STEP-1__ : $\rightarrow$ __INITIALIZATION STEP__
    - Randomly pick one centroid.
    - Calculate __`k-1`__ centroids in such a way that they are far away from each other.
    

- __STEP-2__ : $\rightarrow$ __ASSIGNMENT STEP__
    - Assign each datapoints to its nearest Centroid.
    
    
- __STEP-3__ : $\rightarrow$ __CALCULATION STEP__
    - Recalculate the Centroids.
    $$ \large Centroid_k = \frac{\sum_{x_i  \epsilon \ set_k}  x_i}{|set_k|}$$
    
- __STEP-4__ : $\rightarrow$ __REPETATION STEP__
    - Repeat `STEP-2` & `STEP-3` until convergence.

## <font color="green">NOTE?
- __Assignment Step__ is the most computationally complex(expensive step & k-means has multiple iterations before convergence).

    
- In K-means++, we use a better centroid initialization step which reduces the no. of iterations. Hence reducing total computational complexity of clustering task.

    
- __Assignment Step__ $\Rightarrow$ $ K*n $ distances will be calculated in each iterations.

## <font color="dark blue">ELBOW METHOD
- To find optimal number of Cluster, we use __Elbow Method__.
- It is graphical representation of finding Optimal cluster within a K-Means Clustering.
- In the x-axis the Number of clusters(k value) & y-axis WCSS (Within Clusters Sum of Squared)

## <font color="blue">WCSS
- WCSS stands for __Within Clusters Sum of Squared__.
- WCSS is a measure of how well a clustering algorithm has performed. 
- It calculates the sum of squared distances of each data point to its respective cluster centroid.

## <font color="purple">PARAMETERS OF K-MEANS CLASS

__1. n_clusters=8__:
- This specifies the number of clusters you want to form. In this case, the algorithm will group the data into 8 distinct clusters.


__2. init='k-means++'__:
- This specifies the initialization method for the cluster centroids. 'k-means++' is a smart initialization technique that helps avoid poor initializations, leading to better clustering results.


__3. n_init='auto'__:
- This controls the number of times the K-Means algorithm will be run with different initializations. The final results will be the best run among these.


__4. max_iter=300__:
- This sets the maximum number of iterations the algorithm will run. If the algorithm doesn't converge within 300 iterations, it will stop.


__5. tol=0.0001__:
- This sets the tolerance for the convergence of the algorithm. If the change in the centroids is less than this tolerance, the algorithm stops.


__6. verbose=0__:
- This controls the verbosity of the output. A higher value will print more information about the algorithm's progress.


__7. random_state=None__:
- This sets the random seed for reproducibility. If you set a specific value, you'll get the same results every time you run the algorithm with the same data.


__8. copy_x=True__:
- This controls whether to copy the input data before fitting the model. This can be useful to avoid side effects.


__9. algorithm='lloyd'__:
- This specifies the algorithm to use for optimization. `lloyd` is a classic K-Means algorithm
