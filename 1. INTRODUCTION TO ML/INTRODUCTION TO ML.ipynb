## <font color="blue"> BATCH MACHINE LEARNING<font color/>
- Batch Learning is the conventional way of training machine learning in which we use the entire data to train machine learning model.
- We do it offline. Once the model is trained, then deploy it into the server.
- Batch ML represents the training of ML models in a batch manner.
- Batch ML represents the training of the models at regular intervals such as weekly, monthly, quarterly.

### Disadvantages:
- Lots of data.
- Hardware Limitation
- Availability

### <font color="green"> WHY WE CAN CHOOSE BATCH MACHINE LEARNING?<font color/>
- The bussiness requirements do not require frequent learning of models.
- The data distribution is not expressed to change frequently. Therefore Batch Learning is suitable.
- The software systems required for real-time learning is not available due to various reasons including the cost.
- The expertise required for creating the system for incremental learning is not available.

## <font color="blue"> ONLINE MACHINE LEARNING<font color/>
- In Online ML, the training happens in an incremental manner by continuosly feeding data as it arrives or in a small group/mini batchs.
- Every learning step is fast & cheap, so the system can learn about new data on the fly as it arrives.
- Online learning algorithms can also be used ti train systems on huge datasets that can not fit in one machine's main memory. This is called as __Out-of-core-learning__.

### Learning Rate:
- The rate at which we want our ml to adapt to new dataset is called as __Learning Rate__.
- A system with a high learning rate will tend to forget the learning quickly.
- A system with a low learning rate will be more like __Batch Learning__.

### When to use?
- Where there is a concept of Drift.
- Cost effective.
- Faster solution.

## <font color="blue"> INSTANCE vs MODEL BASED LEARNING<font color/>
- Learning is of two types,<br>
    a. Memorizing (Instance Based Learning) <br>
    b. Generalizing (Model Based Learning) <br>
### <font color="orange">INSTANCE BASED LEARNING<font color/>
- It is also known as __Memory Based Learning__ or __Lazy Learning__.
- It involves memorizing training data in order to make predictions about the future upcoming data points.
- It can be computationally expensive since all of the training data needs to be stored in memory before making a prediction.
- At the time of making prediction, the system uses similarity measure & compare the new cases with learned data.
- eg: KNN
### <font color="orange">MODEL BASED LEARNING<font color/>
- It is also known as __Structure Bases Learning__ or __Eager Learning__.
- It find the __`PATTERN`__ from the training data.
- eg: Linear Regression, Decision Tree, Random Forest etc.
