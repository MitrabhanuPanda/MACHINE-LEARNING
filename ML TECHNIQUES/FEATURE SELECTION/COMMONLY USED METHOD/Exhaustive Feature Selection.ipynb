### 

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector  # Import Exhaustive Feature Selection 
```

```python
efs=ExhaustiveFeatureSelector(RandomForestClassifier(n_jobs=-1),           # All the courses availabe in your pc.
                                                   min_features=1,         # How many minimum feature you want to fit the model.    
                                                   max_features=5,         # How many maximum feature you want to fit the model.
                                                   scoring="accuracy",  
                                                   cv=2,
                                                   n_jobs=-1).fit(x_train,y_train)
```
